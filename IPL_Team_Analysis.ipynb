```python
#Toss Decision
temp_series = df.decision.value_counts()
labels = (np.array(temp_series.index))
sizes = (np.array((temp_series/temp_series.sum())*100))
colors = ['Pink','SkyBlue']
plt.pie(sizes,labels = labels,colors=colors,autopct='%1.1f%%',shadow=True,startangle=90)
plt.title("Toss decision Percentage")
plt.show()
```

```python
#Toss decision Per Season
plt.figure(figsize=(12,6))
sns.countplot(x='season',hue='decision',data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Top players of the match
temp_series = df.pom.value_counts()[:10]
print(temp_series)
labels = np.array(temp_series.index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots()
rects = ax.bar(ind, np.array(temp_series), width=width, color='y')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top player of the match awardees")
plt.show()
```

#**IPL Batting Analysis**

```python
from google.colab import files
data=files.upload()
```

```python
df=pd.read_csv('all_season_batting_card.csv')
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
#Most Run Scored by IPL Teams
df.groupby(['home_team'])['runs'].sum().sort_values(ascending=False)
```

```python
print("Number of matches played: ",df.shape[0])
print("Number of seasons: ",len(df.season.unique()))
```

```python
#Number of wins/team
plt.figure(figsize=(12,6))
sns.countplot(x='current_innings',data=df)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Most IPL Runs by a Batsman
plt.figure(figsize=(10,8))
data = df.groupby(['name'])['runs'].sum().sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

```python
#Avg Run by Teams in Powerplay
df[df['runningOver']<6].groupby(['match_id','home_team']).sum()['runs'].groupby('home_team').mean().sort_values(ascending=False)[2:]
```

```python
#Most IPL Century by a Player
runs = df.groupby(['name','match_id'])['runs'].sum()
runs[runs >= 100].droplevel(level=1).groupby('name').count().sort_values(ascending=False)[:10]
```

```python
#Orange Cap Holder Each Season
data = df.groupby(['season','name'])['runs'].sum().groupby('season').max()
temp_df=pd.DataFrame(df.groupby(['season','name'])['runs'].sum())
print("{0:10}{1:20}{2:30}".format("Season","Player","Runs"))
for season,run in data.items():
    player = temp_df.loc[season][temp_df.loc[season]['runs'] == run].index[0]
    print(season,'\t ',player,'\t\t',run)
```

#**Bowling Statistics**

```python
from google.colab import files
data=files.upload()
```

```python
df=pd.read_csv('all_season_bowling_card.csv')
df
```

```python
df.head(5)
```

```python
# Most run conceded by a bowler in an inning
df.groupby(['name','innings_id'])['conceded'].sum().droplevel(level=1).sort_values(ascending=False)[:10]
```

```python
#Most Dot Ball by a Bowler
plt.figure(figsize=(10,8))
data = df[df['noballs'] == 0].groupby('name').count()['match_id'].sort_values(ascending=False)[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('dots')
plt.ylabel('name')
plt.show()
```

```python
#Most Maiden over by a Bowler
data = df.groupby(['innings_id','name','overs'])['maidens'].sum()
data = data[data.values == 0].droplevel(level=[0,2])
data.index.value_counts()[:10]
```

```python
#Most No Balls by an IPL team
df.groupby(['away_team'])['noballs'].agg('sum').sort_values(ascending=False)
```

```python
#Most No Balls by an IPL Bowler
df[df['noballs'] != 0]['name'].value_counts()[:10]
```

```python
#Most Wides Conceded by an IPL team
df.groupby(['away_team'])['wides'].agg('sum').sort_values(ascending=False)
```
